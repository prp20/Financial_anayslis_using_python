```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as BS
import requests as req
from urllib.request import urlopen, Request
from time import sleep
```

```python
def load_page(link):
    hdr = {'User-Agent': 'Mozilla/5.0'}
    req = Request(link, headers=hdr)
    page = urlopen(req)
    soup = BS(page, features='lxml')
    return soup
```

```python
def scrape_headlines(url):
    soup = load_page(url)
    div_html = soup.find('div', {'class': 'fleft'})
    new_html = div_html.find_all('h2')
    for val in new_html:
        url = val.find('a', href=True)
        news_dict[val.get_text()] = url['href']
```

```python
news_dict = {}
for i in range(1,31):
    sleep(5)
    url = f"https://www.moneycontrol.com/news/business/markets/page-{i}"
    scrape_headlines(url)
    
```

```python
news_data = pd.DataFrame(news_dict.items(), columns=['Headlines', 'Link to Article'])
```
